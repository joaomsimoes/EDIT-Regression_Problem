## Data Cleaning

```python
cols = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']

df = pd.read_csv('imports-85.data', header=None)
df.columns = cols
df
```

```python
df = df.replace('?', np.nan)
```

```python
df.info()
```

```python
print('num-of-cylinders:\n',df['num-of-cylinders'].value_counts(), '\n')
print('num-of-doors:\n',df['num-of-doors'].value_counts())
```

```python
# remove rows without price (price is our label)
df = df[df['price'].notna()]

# convert to numeric values
replace_map_doors = {'four': 4, 'two': 2}
df['num-of-doors'] = df['num-of-doors'].replace(replace_map_doors).astype(float)

replace_map_ncylinders = {'four': 4, 'six': 6, 'five': 5, 'eight': 8, 'two': 2, 'twelve': 12, 'three': 3}
df['num-of-cylinders'] = df['num-of-cylinders'].replace(replace_map_ncylinders).astype(int)

obj_to_int = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
df[obj_to_int] = df[obj_to_int].astype(float)
```

## Prepare for ML

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# separate the predictors and the labels
auto = train_set.drop('price', axis=1)
auto_label = train_set['price'].copy()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

cols_to_impute = ['normalized-losses', 'engine-size', 'curb-weight', 'horsepower', 'width', 'num-of-cylinders', 'length', 'wheel-base', 'bore']

auto_num = auto[cols_to_impute]

imputer.fit(auto_num)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(auto_num)

auto_tr = pd.DataFrame(X, columns=auto_num.columns, index=auto_num.index)
```

```python
from sklearn.preprocessing import OneHotEncoder

cat_attribs = ['engine-location', 'make', 'fuel-type', 'body-style', 'drive-wheels', 'engine-type', 'fuel-system']

auto_cat = auto[cat_attribs]

cat_encoder = OneHotEncoder()
auto_cat_encoded = cat_encoder.fit_transform(auto_cat)
auto_cat_encoded.toarray()
```

```python
cat_encoder.categories_
```

### Pipeline and Feature Scaling

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')), ('std_scaler', StandardScaler())])

auto_num_tr = num_pipeline.fit_transform(auto_num)
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(auto_num)
cat_attribs = ['engine-location', 'make', 'fuel-type', 'body-style', 'drive-wheels', 'engine-type', 'fuel-system']

full_pipeline = ColumnTransformer([('num', num_pipeline, num_attribs), ('cat', OneHotEncoder(), cat_attribs)])
auto_prepared = full_pipeline.fit_transform(auto)
```

```python
auto_prepared
```

### Select and train a model

```python
# LinearRegression

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(auto_prepared, auto_label)
```

```python
from sklearn.metrics import mean_squared_error

auto_pred = lin_reg.predict(auto_prepared)
lin_mse = mean_squared_error(auto_label, auto_pred)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
# RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(auto_prepared, auto_label)
```

```python
auto_predictions = forest_reg.predict(auto_prepared)
lin_mse = mean_squared_error(auto_label, auto_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = [
              {'n_estimators': [80, 100, 120, 140]},
              {'bootstrap': [False]}
]

forest_reg2 = RandomForestRegressor()

grid_search = RandomizedSearchCV(forest_reg2, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(auto_prepared, auto_label)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
  print(np.sqrt(-mean_score), params)
```

```python
final_model = grid_search.best_estimator_

X_test = test_set.drop('price', axis=1)
y_test = test_set['price'].copy()

X_test_prepared = full_pipeline.transform(X_test)

final_predictions = final_model.predict(X_test_prepared)

final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```
